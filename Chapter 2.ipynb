<h4> Exercise 1 </h4>

```python
def posterior(theta):
    return sum((ss.binom.pmf(y,10,theta) * theta**3 * (1-theta)**3
                for y in range(3)))
```

```python
theta_list = np.arange(0,1,0.01)
post_list = [posterior(theta) for theta in theta_list]

plt.plot(theta_list,post_list)
plt.show()
```
